```python
with open(os.path.realpath('input/members_mapping.csv'), 'rb') as f:
    deputats_mapping = pd.read_csv(f) 
```

```python
def request_dep_info(data):
    list_of_dep = []
    for i in deputats_mapping['id']:        
        id_dep = 'id=' + str(i)
        inf = request_url(id_dep, req='deputy')
        if inf.json():
            list_of_dep.append(inf.json())
    return list_of_dep
```

```python
dep_info = request_dep_info(deputats_mapping)
```

```python
dep_info_extracted = pd.io.json.json_normalize(dep_info)
```
